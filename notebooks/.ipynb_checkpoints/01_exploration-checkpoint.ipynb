# ðŸ“Š Exploration du dataset Fashion MNIST

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import fashion_mnist

label_map = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
# repartition viz
plt.figure(figsize=(10, 6))
sns.countplot(x=y_train)
plt.title("RÃ©partition des classes")
plt.xlabel("Classes")
plt.ylabel("Nombre d'images")
plt.xticks(ticks=range(10), labels=label_map, rotation=45)
plt.show()
```

```python
# display samples
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title(label_map[y_train[i]])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
